```python
from sklearn import metrics
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wattermelon_dataset = pd.read_csv("PredictionResultv2.csv")
```

```python
dataset = np.array(wattermelon_dataset)
```

```python
y_true = dataset[:, 1:2]
y_pred = dataset[:, 2:3]
y_score = dataset[:, 3:4]
```

```python
confusion_matrix = metrics.confusion_matrix(y_true=y_true, y_pred=y_pred)
confusion_matrix
```

```python
accuracy = metrics.accuracy_score(y_true=y_true, y_pred=y_pred)
percision_score = metrics.precision_score(y_true=y_true, y_pred=y_pred, pos_label="A")
recall_score = metrics.recall_score(y_true=y_true, y_pred=y_pred, pos_label="A")
f1_score = metrics.f1_score(y_true=y_true, y_pred=y_pred, pos_label="A")
roc_curve = metrics.roc_curve(y_true=y_true, y_score=y_score, pos_label="A")
auc_curve = metrics.auc(roc_curve[0], roc_curve[1])
```

```python
print("Accuracy: ", accuracy)
print("Percision Score: ", percision_score)
print("Recall Score: ", recall_score)
print("F1 Score: ", f1_score)
```

```python
#draw confusion matrix
sns.heatmap(confusion_matrix, annot=True)
plt.show()
```

```python
#draw roc curve
plt.figure()
lw = 2
plt.plot(roc_curve[0], roc_curve[1], color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % auc_curve)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
